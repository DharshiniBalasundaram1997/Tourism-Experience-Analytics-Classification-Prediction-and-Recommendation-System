# **Feature Engineering:**

â– 	Encode categorical variables such as VisitMode, Continent, Country, and AttractionTypeId

```python
#Before Label Encoder
print(User['ContinentId'].unique())
print(User['CountryId'].unique())
print(User['CityId'].unique())
print(User['RegionId'].unique())

print(City['CityId'].unique())
print(City['CountryId'].unique())

print(Country['CountryId'].unique())
print(Country['RegionId'].unique())

print(Region['RegionId'].unique())
print(Region['ContinentId'].unique())

print(Continent['ContinentId'].unique())

print(Type['AttractionTypeId'].unique())

print(Mode['VisitModeId'].unique())

print(Item['AttractionId'].unique())
print(Item['AttractionCityId'].unique())
print(Item['AttractionTypeId'].unique())

print(Transaction['VisitModeId'].unique())
print(Transaction['AttractionId'].unique())
```

```python
label_encoder = LabelEncoder()

# Encode Continent, Region, Country, and City in User
User['ContinentId'] = label_encoder.fit_transform(User['ContinentId'])
User['RegionId'] = label_encoder.fit_transform(User['RegionId'])
User['CountryId'] = label_encoder.fit_transform(User['CountryId'])
User['CityId'] = label_encoder.fit_transform(User['CityId'])

# Encode AttractionTypeId in Mode
City['CityId'] = label_encoder.fit_transform(City['CityId'])
City['CountryId'] = label_encoder.fit_transform(City['CountryId'])

# Encode AttractionTypeId in Mode
Country['CountryId'] = label_encoder.fit_transform(Country['CountryId'])
Country['RegionId'] = label_encoder.fit_transform(Country['RegionId'])

# Encode AttractionTypeId in Mode
Region['ContinentId'] = label_encoder.fit_transform(Region['ContinentId'])
Region['RegionId'] = label_encoder.fit_transform(Region['RegionId'])


Continent['ContinentId'] = label_encoder.fit_transform(Continent['ContinentId'])

# Encode AttractionTypeId in Type
Type['AttractionTypeId'] = label_encoder.fit_transform(Type['AttractionTypeId'])

# Encode AttractionTypeId in Mode
Mode['VisitModeId'] = label_encoder.fit_transform(Mode['VisitModeId'])

# Encode AttractionTypeId in Item
Item['AttractionTypeId'] = label_encoder.fit_transform(Item['AttractionTypeId'])
Item['AttractionCityId'] = label_encoder.fit_transform(Item['AttractionCityId'])
Item['AttractionId'] = label_encoder.fit_transform(Item['AttractionId'])

# Encode VisitMode in Transaction
Transaction['VisitModeId'] = label_encoder.fit_transform(Transaction['VisitModeId'])
Transaction['AttractionId'] = label_encoder.fit_transform(Transaction['AttractionId'])
```

```python
#After Label Encoder
print(User['ContinentId'].unique())
print(User['CountryId'].unique())
print(User['CityId'].unique())
print(User['RegionId'].unique())

print(City['CityId'].unique())
print(City['CountryId'].unique())

print(Country['CountryId'].unique())
print(Country['RegionId'].unique())

print(Region['RegionId'].unique())
print(Region['ContinentId'].unique())

print(Continent['ContinentId'].unique())

print(Type['AttractionTypeId'].unique())

print(Mode['VisitModeId'].unique())

print(Item['AttractionId'].unique())
print(Item['AttractionCityId'].unique())
print(Item['AttractionTypeId'].unique())

print(Transaction['VisitModeId'].unique())
print(Transaction['AttractionId'].unique())
```

â– 	Aggregate user-level features to represent each user's profile (e.g., average ratings per visit mode).

```python
# Average rating per user
user_avg_rating = Transaction.groupby('UserId')['Rating'].mean().reset_index().rename(columns={'Rating': 'Average rating per user'})
user_avg_rating.head()
```

```python
# Count of visits per VisitMode per user
visit_mode_count = Transaction.groupby(['UserId', 'VisitModeId']).size().reset_index()
visit_mode_count

visit_mode_count.columns = ['UserId', 'Visit ModeId', 'Total visits by VisitMode per user']
visit_mode_count = visit_mode_count.drop('Visit ModeId', axis= 1)
visit_mode_count
```

```python
# Average rating per attraction
attraction_avg_rating = Transaction.groupby('AttractionId')['Rating'].mean().reset_index()
attraction_avg_rating

attraction_avg_rating.columns = ['AttractionId', 'Average rating per attraction']
attraction_avg_rating.head()
```

```python
# Total visits per user
user_total_visits = Transaction.groupby('UserId')['TransactionId'].count().reset_index()
user_total_visits

user_total_visits.columns = ['UserId', 'Total visits per user']
user_total_visits
```

```python
Transaction
```

```python
# Merge with Transaction table
Transaction = pd.merge(Transaction,user_avg_rating, on='UserId', how='left')
Transaction = pd.merge(Transaction,visit_mode_count, on='UserId', how='left')
Transaction = pd.merge(Transaction,attraction_avg_rating, on='AttractionId', how='left')
Transaction = pd.merge(Transaction,user_total_visits, on='UserId', how='left')
```
